```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

### Barplot

```python
sns.barplot(data=tips, y="total_bill", x="sex", estimator="max")
```

```python
sns.barplot(data=tips, x="day",y="total_bill", estimator="mean", hue="sex",palette='Set1')
```

```python
sns.barplot(data=tips, x="size", y="total_bill", hue="sex")
```

### Boxplot

```python
sns.boxplot(data=tips, x="day", y="total_bill")
```

### Violinplot

```python
sns.violinplot(data=tips,x="day",y="total_bill",hue="smoker")
```

```python
sns.violinplot(data=tips,x="day",y="total_bill",hue="smoker",split=True)
```

### Swarmplot

```python
sns.swarmplot(x="day",y="total_bill",data=tips)
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips,hue="time")
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips,hue="smoker")
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips,hue="smoker",dodge=True)
```

```python
sns.violinplot(x="tip", y="day", data=tips,palette='rainbow')
sns.swarmplot(x="tip", y="day", data=tips,color='black',size=3)
```

### Stripplot

```python
sns.stripplot(data=tips,x="day",y="total_bill")
```

```python
sns.stripplot(data=tips,x="day",y="total_bill",jitter=False)
```

```python
sns.stripplot(data=tips,x="day",y="total_bill",hue='smoker')
```

```python
sns.stripplot(data=tips,x="day",y="total_bill",hue='smoker',dodge=True)
```

```python
sns.stripplot(data=tips,x="day",y="total_bill",hue='smoker',dodge=True,jitter=False)
```
